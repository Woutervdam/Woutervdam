**Statement:**
- In the game of sticks there is a heap of sticks on a board. On their turn, each player **(two players the computer and you)** picks up a number of sticks. The one who has to pick the final stick will be the loser.
To do this, a number of maximum pick is designated by the player.

- At the beginning of the game, you have to initialize:
  1. The number of sticks between the two players (from 10 to 60).
  2. The maximum number of sticks that can be picked.
  3. Who starts (0 for the player and 1 for the computer).


After each game the program displays the remaining sticks, for example:






```python
import random

def __init__():
    while True:
        try:
            num_sticks = int(input("Enter the number of sticks (10-60): "))
            if 10 <= num_sticks <= 60:
                break
            else:
                print("Please enter a number between 10 and 60.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    while True:
        try:
            max_pick = int(input("Enter the maximum number of sticks that can be picked: "))
            if max_pick > 0:
                break
            else:
                print("Please enter a positive number.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    while True:
        try:
            player_starts = int(input("Who starts? (0 for player, 1 for computer): "))
            if player_starts in (0, 1):
                break
            else:
                print("Please enter either 0 or 1.")
        except ValueError:
            print("Invalid input. Please enter either 0 or 1.")

    return num_sticks, max_pick, player_starts

def player_turn(num_sticks, max_pick):
    while True:
        try:
            num_to_pick = int(input(f"How many sticks do you want to pick (1-{max_pick})? "))
            if 1 <= num_to_pick <= max_pick:
                break
            else:
                print(f"Please enter a number between 1 and {max_pick}.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    
    return num_to_pick

def computer_turn(num_sticks, max_pick):
    num_to_pick = random.randint(1, max_pick)
    print(f"The computer picks {num_to_pick} sticks.")
    return num_to_pick

def display_sticks(n):
        print("o  "*n)
        print("|  "*n)
        
def main():
    num_sticks, max_pick, player_starts = __init__()
    
    while num_sticks > 0:
        if player_starts == 0:
            num_to_pick = player_turn(num_sticks, max_pick)
            print(f"The player picks {num_to_pick} sticks.")
        else:
            num_to_pick = computer_turn(num_sticks, max_pick)
        
        num_sticks =num_sticks - num_to_pick
        print(f"\nNumber of sticks left: {num_sticks}")
        display_sticks(num_sticks)
        player_starts = 1 - player_starts
        
    if player_starts == 0:
        print("\nYou win!")
    else:
        print("\nGame over. Computer won ;(")
```
